```python
# Importando os módulos
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pydot
import graphviz
import os
```

```python
#path para acesso ao exe do graphiz
os.environ["PATH"] += os.pathsep + '/Users/diogo/Anaconda3/pkgs/graphviz-2.38-hfd603c8_2/Library/bin/graphviz'
```

```python
df = pd.read_csv('automobile_data.csv')
```

```python
#movendo a variável target para a extrema direita do dataset
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type',
       'num-of-doors', 'body-style', 'drive-wheels', 'engine-location',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg', 'price', 'aspiration']

df = df.reindex(columns=column_names)
```

```python
#separando variáveis preditoras da target
X = df.iloc[:,:-1]
y = df['aspiration'].values
```

```python
#loop que percorre lista de colunas categóricas que terão valores por categoria convertidas em número 
X = X.astype(str)
novo_valor_numerico = 0
i = 0
for x in X.columns:
    for z in X[x].drop_duplicates():        
        X[x].replace({z: str(i)}, inplace=True)
        i+=1
    i = 0
X = X.astype(int)
```

```python
# Divisão em dados de treino e de teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Construindo o Classificador Baseado em Entropia
# http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
modelo_v1 = DecisionTreeClassifier(max_depth = None, 
                                   max_features = None, 
                                   criterion = 'entropy', 
                                   min_samples_leaf = 1, 
                                   min_samples_split = 2)
```

```python
# Apresentando os dados ao Classificador
modelo_v1.fit(X_train, y_train)
```

```python
# Definindo o nome do arquivo com a árvore de decisão
arquivo = 'tree_modelo_v1.dot'
```

```python
# Gerando o gráfico da árvore de decisão


labels = ['symboling', 'normalized-losses', 'make', 'fuel-type',
       'num-of-doors', 'body-style', 'drive-wheels', 'engine-location',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg', 'price']

export_graphviz(modelo_v1, out_file = arquivo, feature_names = labels)
with open(arquivo) as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```
